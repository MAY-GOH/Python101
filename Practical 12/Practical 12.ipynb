```python
import tkinter as tk # for label frame named in this software tk
from tkinter import * # Import all definitions from tkinter
import tkinter.messagebox

class MonthlySavingsCalculator:
    def __init__(self):
        window = Tk() # Create a window 
        window.title("Monthly $ Calculator") # Set a title
        
        frame1 = Frame(window) # Create and add a frame to window 
        frame1.pack()
        
        label_base = Label(frame1, text = "Base Amount (RM): ")
        self.base = StringVar()
        entry_base = Entry(frame1, textvariable = self.base, width = 20) 

        label_annual_interest = Label(frame1, text = "Annual Interest (%): ")
        self.annual_interest = StringVar()
        entry_annual_interest = Entry(frame1, textvariable = self.annual_interest, width = 20) 

        label_months = Label(frame1, text = "No. of Months (Integer): ")
        self.months = StringVar()
        entry_months = Entry(frame1, textvariable = self.months, width = 20) 
        
        btCalculate = Button(frame1, text = "Calculate!", command = self.processButton)
            
        label_base.grid(row = 1, column = 1, padx = (5, 1), pady = (5, 1), sticky=E) # grid manager
        entry_base.grid(row = 1, column = 2, padx = (1, 5), pady = (5, 1))
        label_annual_interest.grid(row = 2, column = 1, padx = (5, 1), pady = (5, 1), sticky=E)
        entry_annual_interest.grid(row = 2, column = 2, padx = (1, 5), pady = (5, 1))
        label_months.grid(row = 3, column = 1, padx = (5, 1), pady = (5, 1), sticky=E)
        entry_months.grid(row = 3, column = 2, padx = (1, 5), pady = (5, 1))
        
        btCalculate.grid(row = 4, column = 1, columnspan = 2, padx = (5, 5), pady = (5, 5), sticky = W+E)

        frame2 = Frame(window) # Create and add a frame to window 
        frame2.pack()
        
        # Set up a scrollbar object
        scrollV = tk.Scrollbar(frame2, orient=tk.VERTICAL)
        scrollV.pack(side=tk.RIGHT, fill = tk.Y)
        
        # Create and add text to the frame2
        self.text = Text(frame2, height = 10, width = 35, yscrollcommand = scrollV.set) 
        self.text.pack(side = BOTTOM, pady = (1, 10))
        scrollV.config(command = self.text.yview)        
        
        window.mainloop() # Create an event loop     
    
    def processButton(self):
        self.clear()
        self.calculateSavings()
        
    def clear(self):
        self.text.delete('1.0', END)
        
    def calculateSavings(self):
        try:
            monthly_interest = float(self.annual_interest.get()) / (12 * 100)
            base = float(self.base.get())
            months = int(self.months.get())
            
            if monthly_interest < 0 or base < 0 or months < 1:
                tkinter.messagebox.showerror("Error", "Only positive value is allowed!")
                return            
        except:
            tkinter.messagebox.showerror("Error", "Invalid value!")
        else:
            total = 0

            self.text.insert(END, "\t{:10s}{:10s}\n".format("Month", "Total"))
            for count in range(months):
                total = (base + total) * (1 + monthly_interest)

                match (count + 1) % 10:
                    case 1:
                        month = str(count + 1) + "st"
                    case 2:
                        month = str(count + 1) + "nd"
                    case 3: 
                        month = str(count + 1) + "rd"
                    case _:
                        month = str(count + 1) + "th"

                self.text.insert(END, "\t{:10s}RM {:<10.2f}\n".format(month, total))
        
MonthlySavingsCalculator() 

# https://www.homeandlearn.uk/tkinter-scrollbars.html
```

```python
import tkinter.simpledialog
from tkinter import *

def processScores(size):
    scores = []
    counter = 1
    
    while counter <= size:
        message = "Enter the score of the {}{} student".format(str(counter), generateSuffix(counter))
        
        score = tkinter.simpledialog.askfloat("Input", message)
        if score < 0 or score > 100:
            tkinter.messagebox.showerror("Error", "Only positive value in between 0 and 100 is allowed!")
            continue
        
        scores.append(score)
        counter += 1

    # Output: Print average score
    output = "{:s}: {:.0f}%\n".format("Highest", getHighest(scores))
    output += "{:s}: {:.0f}%\n".format("Second Highest", getSecondHighest(scores))
    output += "{:s}: {:.1f}%\n".format("Average", calculateAverage(scores)) 

    tkinter.messagebox.showinfo("Output", output)

def generateSuffix(count):
    match (count % 10):
        case 1:
            return "st"
        case 2: 
            return "nd"
        case 3:
            return "rd"
        case _:
            return "th"

def calculateAverage(scores):
    return sum(scores) / len(scores)
        
def getHighest(scores):
    score_v2 = sorted(scores, reverse=True)
    return score_v2[0]

def getSecondHighest(scores):
    scores_v2 = sorted(scores, reverse=True)
    return scores_v2[1]

def main():
    while True:
        size = tkinter.simpledialog.askinteger("Input", "Enter the number of students")

        if size < 1:
            tkinter.messagebox.showerror("Error", "Only positive value is allowed!")
        else:
            processScores(size)
            break
    
main()
```

### Question \#7B (Using Lists of Objects)

```python
from tkinter import * 
import random

class FrequencyCheckerDemo:
    def __init__(self):
        window = Tk() # Create a window
        window.title("Frequency Checker Demo") # Set title
        
        # frame1 to contain label and canvas
        frame1 = Frame(window)
        frame1.pack()
        
        # Create 10 PhotoImage objects for digit 0 to 9
        file_names = ["0.gif", "1.gif", "2.gif", "3.gif", "4.gif", "5.gif", "6.gif", "7.gif", "8.gif", "9.gif"]
        dir_name = "image"
        image = [PhotoImage(file = (dir_name + "/" + file_names[i])) for i in range(len(file_names))]
        
        # Create 10 Label objects to indicate which digit has the highest frequency
        self.lbl_maxfreq = [Label(frame1, height = 9, width = 10) for i in range(len(image))]
        for i in range(len(image)):
            self.lbl_maxfreq[i].grid(row = 1, column = (i + 1))
        
        # Create 10 Label objects to contain and display digits from 0 to 9 
        lbl = [Label(frame1, image = image[i], height = 170, width = 130, bg="white") for i in range(len(image))]
        for i in range(len(lbl)):
            lbl[i].grid(row = 2, column = (i + 1))
        
        # Create 10 Label objects to display frequency of each digit from 0 to 9
        self.freq = [StringVar() for i in range(len(lbl))]
        entry = [Entry(frame1, text = self.freq[i], width = 5, state = "disabled", justify='center', font = "Calibri 35 bold") for i in range(len(lbl))]
        for i in range(len(entry)):
            entry[i].grid(row = 3, column = (i + 1), pady = (5, 1))
        
        btCalculate = Button(frame1, text = "Generate!", command = self.processButton)
        btCalculate.grid(row = 4, column = 1, columnspan = 10, padx = (5, 5), pady = (5, 5))

        window.mainloop() # Create an event loop

    def processButton(self):
        self.calculateFrequency()
        
    def calculateFrequency(self): 
        self.clearImage()
        
        # Generate 100 random numbers in between 0 and 9 inclusive
        numbers = [random.randint(0, 9) for i in range(100)]
        
        # Calculate the frequency of each digit from 0 to 9
        freq = [0] * 10
        for i in range(len(freq)):
            freq[i] = numbers.count(i)
            self.freq[i].set(freq[i])
        
        # Identify the highest frequency
        maximum = max(freq)
        self.updateImage(freq, maximum)
        
    def updateImage(self, freq, maximum):
        # Create a new image named goodImage to indicate the highest frequecy
        self.goodImage = PhotoImage(file = "image/good.gif")
        
        # Set goodImage to label of every digit with the highest frequecy
        for i in range(len(freq)):
            if freq[i] == maximum:
                self.lbl_maxfreq[i].config(image = self.goodImage, height = 130, width = 130)
                self.lbl_maxfreq[i].image=self.goodImage

    def clearImage(self):
        # Clear all images contained in lbl_maxfreq
        for i in range(len(self.lbl_maxfreq)):
            self.lbl_maxfreq[i].config(image = None)
            self.lbl_maxfreq[i].image=None

FrequencyCheckerDemo() # Create GUI
```
